```python
import abc

# Classe abstrata (Abstração): define a interface comum para tipos de usuário.
class Usuario(abc.ABC):
    def __init__(self, nome):
        # Atributo protegido (Encapsulamento parcial)
        self._nome = nome

    # Método (comportamento da classe)
    @property
    def nome(self):
        return self._nome

    @nome.setter
    def nome(self, valor):
        self._nome = valor

    # Método abstrato — força as subclasses a implementarem (Polimorfismo)
    @abc.abstractmethod
    def get_tipo(self):
        pass

    # Método abstrato — demonstra que diferentes subclasses podem implementar de forma distinta (Polimorfismo)
    @abc.abstractmethod
    def get_desconto(self):
        """Retorna desconto decimal (ex: 0.10 = 10%) — demonstra polimorfismo."""
        pass

    @abc.abstractmethod
    def get_acrescimo(self):
        """Retorna acréscimo decimal (ex: 0.15 = 15%) — demonstra polimorfismo."""
        pass


# Herança: Docente herda de Usuario (Herança)
class Docente(Usuario):
    def __init__(self, nome):
        super().__init__(nome)
        self._tipo = 'docente'

    # Implementação concreta do método abstrato (Polimorfismo)
    def get_tipo(self):
        return self._tipo

    # Implementação concreta do método abstrato (Polimorfismo)
    def get_desconto(self):
        return 0.0  # Sem desconto para docentes

    def get_acrescimo(self):
        return 0.15 # 15% de acréscimo para docentes

# Herança: Discente herda de Usuario (Herança)
class Discente(Usuario):
    def __init__(self, nome):
        super().__init__(nome)
        self._tipo = 'discente'

    def get_tipo(self):
        return self._tipo

    def get_desconto(self):
        return 0.0 # Sem desconto para discentes

    def get_acrescimo(self):
        return 0.0  # Sem acréscimo para discentes

# Classe Marmita: representa um objeto Marmita (Objeto)
# Uso de atributos privados (Encapsulamento) e propriedades para acesso controlado
class Marmita:
    def __init__(self, nome, preco):
        self.__nome = nome       # atributo privado (Encapsulamento)
        self.__preco = float(preco)  # atributo privado (Encapsulamento)

    # Método (getter) para acessar atributo privado
    @property
    def nome(self):
        return self.__nome

    @property
    def preco(self):
        return self.__preco

    # Método especial para exibir a marmita
    def __str__(self):
        return f"{self.nome} - R${self.preco:.2f}"

# Classe Pedido: representa o objeto pedido com seus métodos (Objeto, Métodos)
class Pedido:
    def __init__(self, usuario: Usuario, marmita: Marmita):
        # Atributos protegidos (Encapsulamento)
        self._usuario = usuario
        self._marmita = marmita

    # Método que usa comportamento do usuário (polimorfismo via get_desconto e get_acrescimo)
    def calcular_total(self):
        # Polimorfismo: get_desconto e get_acrescimo se comportam de forma diferente conforme a subclasse (Docente/Discente)
        desconto = self._usuario.get_desconto()
        acrescimo = self._usuario.get_acrescimo()
        return self._marmita.preco * (1 - desconto + acrescimo)

    # Método que resume o pedido (método que combina atributos e comportamentos)
    def resumo(self):
        tipo = self._usuario.get_tipo()
        nome = self._usuario.nome
        desconto_pct = self._usuario.get_desconto() * 100
        acrescimo_pct = self._usuario.get_acrescimo() * 100
        total = self.calcular_total()
        resumo_str = f"Resumo do pedido:\nUsuário: {nome} ({tipo})\nMarmita: {self._marmita.nome} - R${self._marmita.preco:.2f}\n"
        if desconto_pct > 0:
            resumo_str += f"Desconto aplicado: {desconto_pct:.0f}%\n"
        if acrescimo_pct > 0:
            resumo_str += f"Acréscimo aplicado: {acrescimo_pct:.0f}%\n"
        resumo_str += f"Total a pagar: R${total:.2f}"
        return resumo_str


# SistemaMarmitas: classe que gerencia cardápios e busca (Classe com métodos utilitários)
class SistemaMarmitas:
    def __init__(self):
        # cardápio único para ambos os tipos
        self.cardapio = [
            Marmita("Picanha", 25.00),
            Marmita("Bisteca", 20.00),
            Marmita("Alcatra", 22.00),
            Marmita("Bucho", 8.00),
            Marmita("Salsicha", 7.00),
            Marmita("Conserva", 5.00)
        ]

    # Método que imprime o cardápio (Método)
    def mostrar_cardapio(self):
        print("Cardápio:")
        for m in self.cardapio:
            print(f" - {m}")


    # Método que busca uma marmita pelo nome (Método)
    def buscar_marmita(self, nome):
        nome = nome.strip().lower()
        for m in self.cardapio:
            if m.nome.lower() == nome:
                return m
        return None

# Função principal (Função): orquestra a criação de objetos e chamada de métodos
def main():
    sistema = SistemaMarmitas()

    tipo = input("Você é docente ou discente? ").strip().lower()
    if tipo not in ('docente', 'discente'):
        print("Tipo inválido. Digite 'docente' ou 'discente'.")
        return

    # Exibir cardápio imediatamente (interação)
    sistema.mostrar_cardapio()

    nome = input("Digite seu nome: ").strip()
    # Criação de objetos (Instância de classe -> Objeto)
    if tipo == 'docente':
        usuario = Docente(nome)   # Objeto do tipo Docente (Herança + Polimorfismo)
    else:
        usuario = Discente(nome)  # Objeto do tipo Discente (Herança + Polimorfismo)

    escolha = input("Escolha uma marmita pelo nome: ").strip()
    marmita = sistema.buscar_marmita(escolha)
    if marmita is None:
        print("Marmita não encontrada.")
        return

    # Criando objeto Pedido (Objeto) e usando seus métodos
    pedido = Pedido(usuario, marmita)
    print()
    # Exibe resumo do pedido — demonstra uso de atributos e métodos
    print(pedido.resumo())

if __name__ == "__main__":
    main()
    print("Fim do programa.")
```

The subtask requires modifying the `main` function to handle multiple users by adding a `while` loop and a prompt to continue for another user. I will modify the `main` function within the existing code block to incorporate this logic.



```python
import abc

# Classe abstrata (Abstração): define a interface comum para tipos de usuário.
class Usuario(abc.ABC):
    def __init__(self, nome):
        # Atributo protegido (Encapsulamento parcial)
        self._nome = nome

    # Método (comportamento da classe)
    @property
    def nome(self):
        return self._nome

    @nome.setter
    def nome(self, valor):
        self._nome = valor

    # Método abstrato — força as subclasses a implementarem (Polimorfismo)
    @abc.abstractmethod
    def get_tipo(self):
        pass

    # Método abstrato — demonstra que diferentes subclasses podem implementar de forma distinta (Polimorfismo)
    @abc.abstractmethod
    def get_desconto(self):
        """Retorna desconto decimal (ex: 0.10 = 10%) — demonstra polimorfismo."""
        pass

    @abc.abstractmethod
    def get_acrescimo(self):
        """Retorna acréscimo decimal (ex: 0.15 = 15%) — demonstra polimorfismo."""
        pass

# Herança: Docente herda de Usuario (Herança)
class Docente(Usuario):
    def __init__(self, nome):
        super().__init__(nome)
        self._tipo = 'docente'

    # Implementação concreta do método abstrato (Polimorfismo)
    def get_tipo(self):
        return self._tipo

    # Implementação concreta do método abstrato (Polimorfismo)
    def get_desconto(self):
        return 0.0  # Sem desconto para docentes

    def get_acrescimo(self):
        return 0.15 # 15% de acréscimo para docentes

# Herança: Discente herda de Usuario (Herança)
class Discente(Usuario):
    def __init__(self, nome):
        super().__init__(nome)
        self._tipo = 'discente'

    def get_tipo(self):
        return self._tipo

    def get_desconto(self):
        return 0.0 # Sem desconto para discentes

    def get_acrescimo(self):
        return 0.0  # Sem acréscimo para discentes

# Classe Marmita: representa um objeto Marmita (Objeto)
# Uso de atributos privados (Encapsulamento) e propriedades para acesso controlado
class Marmita:
    def __init__(self, nome, preco):
        self.__nome = nome       # atributo privado (Encapsulamento)
        self.__preco = float(preco)  # atributo privado (Encapsulamento)

    # Método (getter) para acessar atributo privado
    @property
    def nome(self):
        return self.__nome

    @property
    def preco(self):
        return self.__preco

    # Método especial para exibir a marmita
    def __str__(self):
        return f"{self.nome} - R${self.preco:.2f}"

# Classe Pedido: representa o objeto pedido com seus métodos (Objeto, Métodos)
class Pedido:
    def __init__(self, usuario: Usuario, marmita: Marmita):
        # Atributos protegidos (Encapsulamento)
        self._usuario = usuario
        self._marmita = marmita

    # Método que usa comportamento do usuário (polimorfismo via get_desconto e get_acrescimo)
    def calcular_total(self):
        if self._usuario.get_tipo() == 'discente':
            return 3.00
        elif self._usuario.get_tipo() == 'docente':
            return 10.00
        else:
            return self._marmita.preco # Default to original price if user type is unknown

    # Método que resume o pedido (método que combina atributos e comportamentos)
    def resumo(self):
        tipo = self._usuario.get_tipo()
        nome = self._usuario.nome
        total = self.calcular_total()
        resumo_str = f"Resumo do pedido:\nUsuário: {nome} ({tipo})\nMarmita: {self._marmita.nome} - R${self._marmita.preco:.2f}\n"
        resumo_str += f"Total a pagar: R${total:.2f}"
        return resumo_str

# SistemaMarmitas: classe que gerencia cardápios e busca (Classe com métodos utilitários)
class SistemaMarmitas:
    def __init__(self):
        # cardápio único para ambos os tipos
        self.cardapio = [
            Marmita("bucho", 8.00),
            Marmita("strogonof", 12.00),
            Marmita("feijão", 6.00),
            Marmita("guizado", 10.00),
            Marmita("porco assado", 15.00)
        ]
        self.vendidas = []
        self.usuarios_ativos = []  # controla os usuários ativos

    # método para adicionar usuário ativo
    def adicionar_usuario(self, usuario):
        self.usuarios_ativos.append(usuario)

    # método para remover usuário ativo
    def remover_usuario(self, usuario):
        if usuario in self.usuarios_ativos:
            self.usuarios_ativos.remove(usuario)

    # Método que imprime o cardápio (Método)
    def mostrar_cardapio(self):
        print("Cardápio:")
        for m in self.cardapio:
            print(f" - {m}")

    # Método que busca uma marmita pelo nome (Método)
    def buscar_marmita(self, nome):
        nome = nome.strip().lower()
        for m in self.cardapio:
            if m.nome.lower() == nome:
                return m
        return None

    def registrar_venda(self, marmita):
        self.vendidas.append(marmita)

    # Só doa marmitas depois que não tem usuários ativos
    def doar_nao_vendidas(self):
        if self.usuarios_ativos:
            print("\nAinda há usuários ativos. As marmitas não podem ser doadas agora.")
            return
        nao_vendidas = [m for m in self.cardapio if m not in self.vendidas]
        if nao_vendidas:
            print("\nMarmitas para doação:")
            for m in nao_vendidas:
                print(f" - {m.nome}")
        else:
            print("\nTodas as marmitas foram vendidas!")


# Função principal (Função): orquestra a criação de objetos e chamada de métodos
def main():
    sistema = SistemaMarmitas()

    while True:
        tipo = input("Você é docente ou discente? ").strip().lower()
        if tipo not in ('docente', 'discente'):
            print("Tipo inválido. Digite 'docente' ou 'discente'.")
            continue # Ask again for valid type

        # Exibir cardápio imediatamente (interação)
        sistema.mostrar_cardapio()

        nome = input("Digite seu nome: ").strip()
        # Criação de objetos (Instância de classe -> Objeto)
        if tipo == 'docente':
            usuario = Docente(nome)   # Objeto do tipo Docente (Herança + Polimorfismo)
        else:
            usuario = Discente(nome)  # Objeto do tipo Discente (Herança + Polimorfismo)

        sistema.adicionar_usuario(usuario)  # adiciona usuário ativo

        escolha = input("Escolha uma marmita pelo nome: ").strip()
        marmita = sistema.buscar_marmita(escolha)
        if marmita is None:
            print("Marmita não encontrada.")
            sistema.remover_usuario(usuario)  # remove usuário que não concluiu compra
            continue # Allow user to choose again or exit

        # Creating Pedido object (Object) and using its methods
        pedido = Pedido(usuario, marmita)
        print()
        # Display order summary - demonstrates use of attributes and methods
        print(pedido.resumo())

        # Register the sold marmita
        sistema.registrar_venda(marmita)

        sistema.remover_usuario(usuario)  # remove usuário ativo após finalizar pedido

        another_user = input("Mais algum usuário quer fazer um pedido? (yes/no): ").strip().lower()
        if another_user == 'no':
            break

    # Doar the unsold marmitas after the loop finishes
    sistema.doar_nao_vendidas()

if __name__ == "__main__":
    main()
    print("Fim do programa.")
```
